```python
#! /usr/env/bin python
# -*- coding: utf-8 -*-


"""
A stem and leaf plot is a way of summarizing a set of data measured on an
interval scale. It is often used in exploratory data analysis to illustrate the
major features of the distribution of the data in a convenient and easily drawn
form.
A stem and leaf plot is similar to a histogram but is usually a more informative
display for relatively small data sets (<100 data points).
It provides a table as well as a picture of the data and from it we can readily
write down the data in order of magnitude, which is useful for many statistical
procedures.
"""


import stemgraphic
import pandas as pd
from jupyterthemes import jtplot
jtplot.style()

x = [35, 42, 38, 57, 2, 24, 27, 36, 45, 60, 38, 40, 
     40, 44, 1, 44, 48, 84, 38, 20, 4, 2, 48, 58, 21]

y = pd.Series(x)
fig, ax = stemgraphic.stem_graphic(y)
```
